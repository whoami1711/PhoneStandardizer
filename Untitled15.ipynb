```python
from stable_baselines3.common.callbacks import BaseCallback

class LagrangianCallbackAdv(BaseCallback):
    def __init__(self, venv, c1=0.02, c2=0.02, eta=0.1, gamma=0.99, verbose=1):
        super().__init__(verbose)
        self.venv = venv
        self.c1, self.c2 = c1, c2
        self.eta, self.gamma = eta, gamma
        self.lambda1 = 0.0
        self.lambda2 = 0.0
        self._reset_roll()

    def _reset_roll(self):
        self.sum_w = 0.0
        self.sum_g1_w = 0.0
        self.sum_g2_w = 0.0
        self.t = 0

    def _on_training_start(self) -> None:
        # Đẩy λ vào model custom
        try:
            self.model.lagrange_lambdas = (self.lambda1, self.lambda2)
        except Exception:
            pass

    def _on_rollout_start(self) -> None:
        self._reset_roll()

    def _on_step(self) -> bool:
        infos = self.locals.get("infos", [])
        w = (self.gamma ** self.t)
        for info in infos:
            if isinstance(info, dict):
                self.sum_g1_w += w * float(info.get("g1", 0.0))
                self.sum_g2_w += w * float(info.get("g2", 0.0))
                self.sum_w += w
                self.t += 1
        return True

    def _on_rollout_end(self) -> None:
        if self.sum_w > 0:
            g1_bar = self.sum_g1_w / self.sum_w
            g2_bar = self.sum_g2_w / self.sum_w
            # Dual ascent
            self.lambda1 = max(0.0, self.lambda1 + self.eta * (g1_bar - self.c1))
            self.lambda2 = max(0.0, self.lambda2 + self.eta * (g2_bar - self.c2))
            if self.verbose:
                print(f"[Dual] λ1={self.lambda1:.4f}, λ2={self.lambda2:.4f} | g1={g1_bar:.4f}, g2={g2_bar:.4f}")
        # Cập nhật vào model trước rollout kế tiếp
        try:
            self.model.lagrange_lambdas = (self.lambda1, self.lambda2)
        except Exception:
            pass
        self._reset_roll()
```

```python
# === Cấu hình chung theo bài báo ===
from agent.pdppo_adv_train import train_adv
from agent.custom_ppo import CustomPPO
from envs.autoscale_env import AutoscaleEnv

# Generator tải đúng Eq.(load) trong bài
env_kwargs = dict(base=50.0, amp=40.0, spike=200.0, sigma=15.0)

# Lưới constraint levels cho Pareto
C1 = [0.01, 0.02, 0.05]
C2 = [0.01, 0.02, 0.05]

# Số bước train (bản nhanh để kiểm thử; khi chạy chính cho bài, tăng lên 200_000)
TOTAL_STEPS = 20_480  # ~20k "warm-up"; final: 200_000

# Danh sách seed (bản nhanh). Khi đánh giá chính thức: dùng 3–5 seed cho train.
SEEDS = [7, 11, 19]

# Shield margin kappa và tail-SLO T thiết lập trong env của bạn (nếu có tham số).
# Ở đây giả sử mặc định đã là T=200ms, kappa=0 (như bài).
```

```python
import os, json, pathlib

runs = []

for shield in [False, True]:
    for c1 in C1:
        for c2 in C2:
            for seed in SEEDS:
                save_dir = f"runs/pdppo_shield{int(shield)}_c1{c1}_c2{c2}_seed{seed}"
                pathlib.Path(save_dir).mkdir(parents=True, exist_ok=True)

                model_path = train_adv(
                    save_dir=save_dir,
                    total_steps=TOTAL_STEPS,
                    seed=seed,
                    gamma_rl=0.99,
                    c1=c1, c2=c2, eta=0.1,
                    shield=shield,
                    env_kwargs=env_kwargs,
                )
                runs.append(dict(
                    method="PD-PPO+shield" if shield else "PD-PPO",
                    shield=shield, c1=c1, c2=c2, seed=seed,
                    model_path=model_path, save_dir=save_dir
                ))

print(f"Đã train {len(runs)} mô hình.")
```

```python
#ham danh gia nang luong PAC
import numpy as np, math

def evaluate_model(model, episodes=50, shield=True, seed=123, env_kwargs=None):
    """Chạy N episode, trả metrics theo bài: p_hat, PAC-UB, avg_P, g2_rate."""
    env_kwargs = env_kwargs or {}
    env = AutoscaleEnv(shield=shield, **env_kwargs)

    results = []
    for ep in range(episodes):
        obs, _ = env.reset(seed=seed + ep)
        violated = 0
        g2_sum = 0.0
        p_sum = 0.0
        steps = 0

        while True:
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, terminated, truncated, info = env.step(action)
            steps += 1
            violated = max(violated, int(info.get("g1", 0)))
            g2_sum += float(info.get("g2", 0.0))
            p_sum  += float(info.get("P", 0.0))
            if terminated or truncated:
                break

        results.append({
            "viol_ep": violated,
            "avg_P": p_sum / max(steps, 1),
            "g2_rate": g2_sum / max(steps, 1),
            "steps": steps
        })

    N = len(results)
    k = int(np.sum([r["viol_ep"] for r in results]))
    p_hat = k / N
    delta = 0.05
    pac_ub = p_hat + math.sqrt((1/(2*N)) * math.log(1/delta))

    avgP = np.array([r["avg_P"] for r in results])
    g2r  = np.array([r["g2_rate"] for r in results])

    summary = {
        "N": N,
        "k": k,
        "p_hat": float(p_hat),
        "pac_ub@0.05": float(pac_ub),
        "avg_P_mean": float(avgP.mean()),
        "avg_P_std": float(avgP.std(ddof=1) if N > 1 else 0.0),
        "g2_rate_mean": float(g2r.mean()),
        "g2_rate_std": float(g2r.std(ddof=1) if N > 1 else 0.0),
    }
    return summary, results
```

```python
import pandas as pd

rows = []
per_episode_dump = []  # nếu muốn lưu chi tiết từng episode

for r in runs:
    model = CustomPPO.load(r["model_path"])
    # Đánh giá chính: N >= 200 theo bài; kiểm thử nhanh đặt 50
    summary, episodes = evaluate_model(
        model,
        episodes=50,          # final: 200
        shield=r["shield"],   # đánh giá đúng cấu hình triển khai
        seed=1000,            # seed khác train
        env_kwargs=env_kwargs
    )

    row = dict(
        method=r["method"], shield=r["shield"],
        c1=r["c1"], c2=r["c2"], seed=r["seed"],
        total_steps=TOTAL_STEPS, **summary
    )
    rows.append(row)

    # (tuỳ chọn) lưu chi tiết từng episode
    for i, ep in enumerate(episodes):
        per_episode_dump.append({
            "method": r["method"], "shield": r["shield"],
            "c1": r["c1"], "c2": r["c2"], "seed": r["seed"], "ep": i,
            **ep
        })

df = pd.DataFrame(rows)
df_ep = pd.DataFrame(per_episode_dump)

df.to_csv("summary_pareto_pac.csv", index=False)
df_ep.to_csv("episodes_detail.csv", index=False)

df.sort_values(["method","c1","c2"]).head()
```
